```python
import pandas as pd

data = pd.read_csv("disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.csv")
df = pd.DataFrame(data)

df
```

```python
# No 2 untuk Tahun 2018
tahun_2018 = 2018

sampah_2018 = 0
for i, row in df.iterrows():
    if row['tahun'] == tahun_2018:
        sampah_2018 += row['jumlah_produksi_sampah']

df['total_sampah_2018'] = sampah_2018

df
```

```python
# No 3 Total sampah per tahun
total_sampah_per_tahun = {tahun: 0 for tahun in range(2015, 2024)}

for i, row in df.iterrows():
    tahun = row['tahun']
    jumlah_sampah = row['jumlah_produksi_sampah']
    if tahun in total_sampah_per_tahun:
        total_sampah_per_tahun[tahun] += jumlah_sampah

total_sampah_per_tahun_df = pd.DataFrame(
    list(total_sampah_per_tahun.items()),
    columns=['tahun', 'total_sampah_per_tahun']
)

print(total_sampah_per_tahun)
total_sampah_per_tahun_df
```

```python
# No 3 Menjumlahkan data per kota per tahun
total_sampah = {}

for i, row in df.iterrows():
    kabupaten_kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    jumlah_sampah = row['jumlah_produksi_sampah']

    kota = (kabupaten_kota, tahun)

    if kota in total_sampah:
        total_sampah[kota] += jumlah_sampah
    else:
        total_sampah[kota] = jumlah_sampah

result = pd.DataFrame(list(total_sampah.items()), columns=['kabupaten_kota_tahun', 'total_sampah_per_kota'])
result[['kabupaten_kota', 'tahun']] = pd.DataFrame(result['kabupaten_kota_tahun'].to_list(), index=result.index)
result.drop(columns=['kabupaten_kota_tahun'], inplace=True)

result
```

```python
df_jawaban = total_sampah_per_tahun_df.copy()
df_jawaban['total_sampah_2018'] = sampah_2018

df_baru = df_jawaban.merge(result, on='tahun', how='left')

print(df_baru)
```

```python
df_baru.to_excel("Hasil Akhir.xlsx", index=False)
df_baru.to_csv("Hasil Akhir.csv", index=False) 
```
